Grid Search and Randomized Search CV Implementation in Machine Learning. 

Here we are going to look at Gird Search CV and Randomized Search CV and find the best parameters for our model.

Here we got 95% accuracy for our model which is really good.

The model we are going to use here is called svm (Support Vector Machine)

After finding our parameters we will give our own input vale and we will let our model make predictions.

```python
import pandas as pd
import numpy as np

import sklearn.datasets
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer
```

```python
breast = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
breast
```

```python
breast['label'] = breast_cancer.target
```

```python
breast
```

```python
X = breast.drop(columns='label').values
```

```python
Y = breast['label']
```

#**Grid Search CV**

```python
model = SVC()
```

```python
parameters = {'kernel':['linear','poly','rbf','sigmoid'], 'C':[1,5,10,20]}
```

```python
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
classifier.fit(X,Y)
```

```python
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
best_score = classifier.best_score_
print(best_score)
```

```python
classifier.cv_results_
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result
```

```python
grid_search_result = result[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
grid_search_result
```

#**Randomized Search CV**

```python
model2 = SVC()
```

```python
parameters2 = {'kernel':['linear', 'poly', 'rbf','sigmoid'], 'C':[1,5,10,20]}
```

```python
classifier2 = RandomizedSearchCV(model2, parameters2, cv=5)
```

```python
classifier2.fit(X,Y)
```

```python
best_parameters2 = classifier2.best_params_
print(best_parameters)
```

```python
best_score2 = classifier2.best_score_
print(best_score2)
```

```python
classifier2.cv_results_
```

```python
results2 = pd.DataFrame(classifier2.cv_results_)
```

```python
results2
```

```python
randomized_search_results = results2[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
randomized_search_results
```

#**Train Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model3 = SVC(kernel='linear', C=10)
```

```python
model3.fit(X_train, Y_train)
```

```python
X_train_prediction = model3.predict(X_train)
X_train_accuracyscore = accuracy_score(X_train_prediction, Y_train)
print(X_train_accuracyscore)
```

```python
X_test_prediction = model3.predict(X_test)
X_test_accuracyscore = accuracy_score(X_test_prediction, Y_test)
print(X_test_accuracyscore)
```
